```python
df = pd.read_csv("main.csv")
df
```

```python
adjusted_rank = df["Adjusted Rank"]
tuition = df["Tuition"]
enrollment = df["Enrollment Numbers"]

# Create a scatter plot
plt.scatter(adjusted_rank, tuition)
plt.xlabel("Adjusted Rank")
plt.ylabel("Tuition in Dollars$")
```

```python

plt.scatter(df['Tuition'], df['Enrollment Numbers'], c=df['Adjusted Rank'], cmap='viridis')
plt.xlabel('Tuition ($)')
plt.ylabel('Enrollment Numbers')
plt.title('Enrollment vs. Tuition')
plt.colorbar(label='Adjusted Rank')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_10 = df.nsmallest(15, 'Adjusted Rank')
plt.bar(top_10['College Name'], top_10['Adjusted Rank'])
plt.xlabel('University')
plt.ylabel('Adjusted Rank')
plt.title('Top 10 Universities by Adjusted Rank')
plt.xticks(rotation=90)
plt.show()

```

```python
top_15 = df.nlargest(10, 'Tuition')
plt.bar(top_15['College Name'], top_15['Tuition'])
plt.xlabel('University')
plt.ylabel('Adjusted Rank')
plt.title('Top 15 Universities by Adjusted Rank')
plt.xticks(rotation=90)
# f = io.StringIO() 
# fig.savefig(f, format="svg")
# plt.close()
```
